```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Read json data

gdf_states_raw = gpd.read_file(r'data\gadm41_BRA_1.json')
gdf_states_raw
```

```python
#Tratament of geo Data and extract colum "HASC_1", also "geometry"

gdf_states_etl = gdf_states_raw.copy()

gdf_states_etl = gdf_states_etl[['HASC_1', 'geometry']]

#gdf_states_etl = gdf_states_etl[gdf_states_etl['HASC_1'] == 'BR.CE']

gdf_states_etl
```

```python
# Tratament of HASC_1 to remove 'BR.' and rename to 'states'

gdf_states_etl['HASC_1'] = gdf_states_etl['HASC_1'].str.replace('BR.', '')
gdf_states_etl = gdf_states_etl.rename(columns = {'HASC_1': 'states'})
gdf_states_etl
```

```python
# Read obits csv data 

gdf_obits_Firearms_raw = gpd.read_file(r'data\bitos-por-armas-de-fogo.csv')
gdf_obits_Firearms_raw
```

```python
#Dropping 'cod' collumn

gdf_obits_Firearms_etl = gdf_obits_Firearms_raw.copy()
gdf_obits_Firearms_etl = gdf_obits_Firearms_etl.drop(columns = ['cod'])
gdf_obits_Firearms_etl
```

```python
''#Rename of column 'nome', 'perÃ­odo' and 'valor' to 'states', 'year' and 'obits', respectively.

gdf_obits_Firearms = gdf_obits_Firearms_etl.copy()
gdf_obits_Firearms = gdf_obits_Firearms.rename(columns = {'nome': 'states','perÃ­odo' : 'year' ,'valor': 'obits'})
gdf_obits_Firearms
```

```python
# Tratament of types of 'obits' and 'yeear' columns. Also join geometry GeoDataFrame with Data Obits/year

gdf_obits_Firearms_general = gdf_obits_Firearms.join(gdf_states_etl.set_index('states'), on='states', how='left')
gdf_obits_Firearms_general['year'] = gdf_obits_Firearms_general['year'].astype('int')
gdf_obits_Firearms_general['obits'] = gdf_obits_Firearms_general['obits'].astype('int')
gdf_obits_Firearms_general
```

```python
# Remove 'year' column

gdf_obits_Firearms_2020 = gdf_obits_Firearms_general.copy()
gdf_obits_Firearms_2020 = gdf_obits_Firearms_2020[gdf_obits_Firearms_2020['year'] == 2020].drop(columns = ['year'])
gdf_obits_Firearms_2020
```

```python
# Sum of obits per year

gdf_obits_Firearms_general_sum = gdf_obits_Firearms_general.copy()

gdf_obits_Firearms_general_sum = gdf_obits_Firearms_general_sum.groupby('states').agg({
    'year': 'first',           # Mantém o primeiro ano, já que é o mesmo para cada estado
    'obits': 'sum',            # Soma o número de óbitos por estado
    'geometry': 'first'        # Mantém a primeira geometria (mesma para cada estado)
}).reset_index().drop(columns = ['year'])

gdf_obits_Firearms_general_sum
```

```python
# Ensuring that the 'geometry' column is a geometry type

gdf_obits_Firearms_general_sum_geometry = gdf_obits_Firearms_general_sum.copy()
gdf_obits_Firearms_general_sum_geometry = gpd.GeoDataFrame(gdf_obits_Firearms_general_sum_geometry, geometry='geometry')
```

```python

# plot geo DataFrame of sum all obits per states

fig, ax = plt.subplots(figsize=(10, 10))
gdf_obits_Firearms_general_sum_geometry.plot(column='obits', cmap='OrRd', legend=True, ax=ax)

ax.set_title('Aggregation of obits per States of Brazil', fontsize=15)
ax.set_axis_off()

plt.show()
```

```python
#Ensuring that the 'geometry' column is a geometry type

gdf_obits_Firearms_2020_geometry = gdf_obits_Firearms_2020.copy()
gdf_obits_Firearms_2020_geometry = gpd.GeoDataFrame(gdf_obits_Firearms_2020_geometry, geometry='geometry')
```

```python
# plot geo DataFrame of Obits of 2020

fig, ax = plt.subplots(figsize=(10, 10))
gdf_obits_Firearms_2020_geometry.plot(column='obits', cmap='OrRd', legend=True, ax=ax)

ax.set_title('Obits per States of Brazil in 2020', fontsize=15)
ax.set_axis_off()

plt.show()
```
